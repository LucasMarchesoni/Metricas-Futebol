# 1. Importação das bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("fivethirtyeight")
pd.set_option('display.max_columns', None)
```

# 2. Importação da base de dados

```python
df_regular_season = pd.read_csv("./arquivos/rodada35/regular_overall_season_serie_a.csv")
df_regular_home_away = pd.read_csv("./arquivos/rodada35/regular_overall_homeaway_serie_a.csv")
```

```python
df_regular_season.head()
```

```python
df_regular_home_away.head()
```

```python
df_regular_season.drop(['Notes'], inplace=True, axis=1)
```

```python
df_regular_season.head()
```

```python
df_regular_season[['Top_Scorer', 'Goals']] = df_regular_season['Top Team Scorer'].str.split('-', 1, expand=True)
df_regular_season.drop(['Top Team Scorer'], inplace=True, axis=1)
df_regular_season.head()
```

```python
df_ordenado_por_vitorias = df_regular_season.sort_values('W', ascending=False)
plt.figure(figsize=(20, 10))
plt.title("Ranking de vitórias")
plt.xlabel("Time")
plt.ylabel("Vitórias")
plt.xticks(rotation=45)
sns.barplot(x = 'Squad', y = 'W', data = df_ordenado_por_vitorias, palette = 'rocket')
```

```python
df_ordenado_por_vitorias_em_casa = df_regular_home_away.sort_values('W', ascending=False)
plt.figure(figsize=(20, 10))
plt.title("Ranking de vitórias em casa")
plt.xlabel("Time")
plt.ylabel("Vitórias")
plt.xticks(rotation=45)
sns.barplot(x = 'Squad', y = 'W', data = df_ordenado_por_vitorias_em_casa, palette = 'rocket')
```

```python
df_ordenado_por_vitorias_fora = df_regular_home_away.sort_values('W.1', ascending=False)
plt.figure(figsize=(20, 10))
plt.title("Ranking de vitórias fora de casa")
plt.xlabel("Time")
plt.ylabel("Vitórias")
plt.xticks(rotation=45)
sns.barplot(x = 'Squad', y = 'W.1', data = df_ordenado_por_vitorias_fora, palette = 'rocket')
```

```python
df_ordenado_por_empates = df_regular_season.sort_values('D', ascending=False)
plt.figure(figsize=(20, 10))
plt.title("Ranking de empates")
plt.xlabel("Time")
plt.ylabel("Empates")
plt.xticks(rotation=45)
sns.barplot(x = 'Squad', y = 'D', data = df_ordenado_por_empates, palette = 'rocket')
```

```python
df_ordenado_por_derrotas = df_regular_season.sort_values('L', ascending=False)
plt.figure(figsize=(20, 10))
plt.title("Ranking de derrotas")
plt.xlabel("Time")
plt.ylabel("Derrotas")
plt.xticks(rotation=45)
sns.barplot(x = 'Squad', y = 'L', data = df_ordenado_por_derrotas, palette = 'rocket')
```

```python
df_ordenado_por_gols_feitos = df_regular_season.sort_values('GF', ascending=False)
plt.figure(figsize=(20, 10))
plt.title("Ranking de gols feitos")
plt.xlabel("Time")
plt.ylabel("Gols Feitos")
plt.xticks(rotation=45)
sns.barplot(x = 'Squad', y = 'GF', data = df_ordenado_por_gols_feitos, palette = 'rocket')
```

```python
df_ordenado_por_gols_tomados = df_regular_season.sort_values('GA', ascending=False)
plt.figure(figsize=(20, 10))
plt.title("Ranking de gols tomados")
plt.xlabel("Time")
plt.ylabel("Gols Tomados")
plt.xticks(rotation=45)
sns.barplot(x = 'Squad', y = 'GA', data = df_ordenado_por_gols_tomados, palette = 'rocket')
```

```python
plt.figure(figsize=(15, 10))
plt.title("Gols feitos esperados X Gols sofridos esperados")
sns.scatterplot(y='xGA', x='xG', data=df_regular_season)
for index, row in df_regular_season.iterrows():
    plt.annotate(row['Squad'], (row['xG'], row['xGA']), textcoords="offset points", xytext=(0,5), ha='center')
plt.gca().invert_yaxis()
```

```python
plt.figure(figsize=(15, 10))
plt.title("Gols feitos esperados em casa X Gols sofridos esperados em casa")
sns.scatterplot(y='xGA', x='xG', data=df_regular_home_away)
for index, row in df_regular_home_away.iterrows():
    plt.annotate(row['Squad'], (row['xG'], row['xGA']), textcoords="offset points", xytext=(0,5), ha='center')
plt.gca().invert_yaxis()
```

```python
plt.figure(figsize=(15, 10))
plt.title("Gols feitos esperados fora de casa X Gols sofridos esperados fora de casa")
plt.xlabel("xG")
plt.ylabel("xGA")
sns.scatterplot(y='xGA.1', x='xG.1', data=df_regular_home_away)
for index, row in df_regular_home_away.iterrows():
    plt.annotate(row['Squad'], (row['xG.1'], row['xGA.1']), textcoords="offset points", xytext=(0,5), ha='center')
plt.gca().invert_yaxis()
```
